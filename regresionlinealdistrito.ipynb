```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import gdown
from google.colab import files

# Función para descargar archivos de Google Drive
def download_file_from_google_drive(drive_url, output_path):
    file_id = drive_url.split('/')[-2]
    gdown.download(f'https://drive.google.com/uc?id=' + file_id, output_path, quiet=False)

# URLs de los archivos en Google Drive
drive_url1 = 'https://docs.google.com/spreadsheets/d/1vUq_bie2H7bWx-65ZmIrE0U80Xp6mnBI/edit?usp=drive_link&ouid=115791221559135717387&rtpof=true&sd=true'

# Descargar archivos de Google Drive
file_path1 = '/home/data_procesada.xlsx'
download_file_from_google_drive(drive_url1, file_path1)

data = pd.read_excel(file_path1)

data['PERIODO'] = data['PERIODO'].astype(int)
data['POB_URBANA'] = data['POB_URBANA'].astype(float)
data['GENERACION_MUN_TANIO'] = data['GENERACION_MUN_TANIO'].astype(float)

proyecciones = []
distritos = data['DISTRITO'].unique()

for distrito in distritos:
    df_distrito = data[data['DISTRITO'] == distrito]
    X = df_distrito[['PERIODO', 'POB_URBANA']]
    y = df_distrito['GENERACION_MUN_TANIO']

    model = LinearRegression()
    model.fit(X, y)

    # Evalua el modelo
    score = model.score(X, y)
    print(f"Modelo para {distrito}: R^2 = {score:.4f}")

    # Proyecta a tres años en el futuro
    ultimo_periodo = df_distrito['PERIODO'].max()
    proximo_periodo_1 = ultimo_periodo + 1
    proximo_periodo_2 = ultimo_periodo + 2
    proximo_periodo_3 = ultimo_periodo + 3

    # La población urbana sigue la misma tendencia de crecimiento
    ultima_poblacion = df_distrito[df_distrito['PERIODO'] == ultimo_periodo]['POB_URBANA'].values[0]
    crecimiento_poblacion = (ultima_poblacion - df_distrito[df_distrito['PERIODO'] == (ultimo_periodo - 1)]['POB_URBANA'].values[0])

    def prediccionResiduos(x):
        diferencia = x - ultimo_periodo
        poblacion_futura = ultima_poblacion + crecimiento_poblacion*diferencia
        prediccion= model.predict([[x, poblacion_futura]])[0]
        proyecciones.append({
            'DISTRITO': distrito,
            'PERIODO': x,
            'GENERACION_MUN_TANIO': prediccion
        })
        return prediccion


    plt.figure(figsize=(10, 6))
    plt.plot(df_distrito['PERIODO'], df_distrito['GENERACION_MUN_TANIO'], label='Histórico')
    plt.plot([proximo_periodo_1, proximo_periodo_2,proximo_periodo_3],
              [prediccionResiduos(proximo_periodo_1), prediccionResiduos(proximo_periodo_2),prediccionResiduos(proximo_periodo_3)],
              linestyle='--', marker='o', color='red', label='Proyección')
    plt.xlabel('Periodo')
    plt.ylabel('GENERACION_MUN_TANIO')
    plt.title(f'Proyección de GENERACION_MUN_TANIO para {distrito}')
    plt.legend()
    plt.grid(True)
    plt.show()

proyecciones_df = pd.DataFrame(proyecciones)

print(proyecciones_df)

proyecciones_df.to_excel('/home/proyecciones.xlsx', index=False)

files.download('/home/proyecciones.xlsx')
```
